## Label Encoding 

```
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import sklearn
from sklearn.preprocessing import LabelEncoder
```

```
from google.colab import files 
df=files.upload()
```

```
import io
df_train=pd.read_csv(io.StringIO(df['train.csv'].decode('utf-8')))
```

```
df_train.head()
```

```
df2=df_train[["KitchenQual","BldgType"]]
```

```
lE=LabelEncoder()
```

```
lE.fit_transform(df2["BldgType"])
```

```
pd.set_option("display.max_rows",None)
```

```
df2["BldgType_L_enc"]=lE.fit_transform(df2["BldgType"])
df2.head()
```

```
df2["BldgType"].value_counts()
```

## Ordinal Encoding

```
df_train["KitchenQual"].value_counts()
```

```
order_label={"Ex":4,"Gd":3,"TA":2,"Fa":1}
```

```
df2["KitchenQual_ord_enc"]=df2["KitchenQual"].map(order_label)
df2.head()
```
