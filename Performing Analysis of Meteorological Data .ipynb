## **Extracting the required data**

# Changing the Formatted Date column into datetime object so that we can use it to extarct out yearly, monthly and daily data to analyse the variation of tem perature and humidity. 

```
new_df["Formatted Date"]= pd.to_datetime(new_df["Formatted Date"],utc=True)
```

Setting the formatted date as index which helps us during the time of plotting data.

```
ndf=new_df.set_index('Formatted Date')
ndf=ndf.resample('MS').mean()
```

Plotting the data i.e., Temperature and Humidity vs. Year using matplotlib.
We can clearly see the mean temperature and humidity remains almost same throughout the 10 years of time period. 

```
plt.figure(figsize=(14,6))
plt.title("Variation of temp and humidity w.r.t. Year")
plt.plot(ndf)
```
