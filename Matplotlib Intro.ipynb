```
import matplotlib.pyplot as plt
plt.plot([2,6,4,5])# If you provide a single list or array to the plot() matplotlib assumes it is a sequence of y values and automatically generate x values for youm
plt.xlabel("Indices")
plt.ylabel("Number")
plt.title('My Plot')
plt.show()
```

```
import numpy as np
t=np.arange(0,5,0.2)
plt.plot(t,t**2,'b--',label='^2')#b-- meant blue dashed line
plt.plot(t,t**2.2,'g^',label='^2.2')#g^ meant green triagle
plt.plot(t,t**2.5,'rs',label='^2.5')#rs meant red square
plt.grid()
plt.legend()#add a table of instructions used in plot
plt.show()
```

# Controlling line Width

```
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines=plt.plot(x1,y1,x2,y2)
plt.grid()
plt.setp(lines[0],color='r',linewidth=2.0)
plt.setp(lines[1],color='g',linewidth=2.0)
```

```
def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)  
t2=np.arange(0.0,5.0,0.2)
plt.figure(1)
plt.subplot(211)#211 mean -> 2 rows, 1 column, figure 1 part 1
plt.grid()
plt.plot(t1,f(t1),'b-')
plt.subplot(212)#212 mean -> 2 rows, 1 column, figure 1 part 2
plt.plot(t2,np.cos(2*np.pi*t),'r--')
plt.show()
```
